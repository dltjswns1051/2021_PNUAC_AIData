```python
import matplotlib.pyplot as plt
import pandas as pd

#   
csv_file_path = '/content/GDP Trend_all country.csv'

# CSV  
df = pd.read_csv(csv_file_path)

#   
asian_countries = [
    'Afghanistan', 'Armenia', 'Azerbaijan', 'Bahrain', 'Bhutan', 'Brunei Darussalam',
    'Bangladesh', 'Cambodia', 'China', 'Cyprus', 'Georgia', 'India', 'Indonesia',
    'Iran, Islamic Rep.', 'Iraq', 'Israel', 'Japan', 'Jordan', 'Kazakhstan',
    'Korea, Rep.', 'Kuwait', 'Kyrgyz Republic', 'Lao PDR', 'Lebanon', 'Malaysia',
    'Maldives', 'Mongolia', 'Myanmar', 'Nepal', 'Oman', 'Pakistan', 'Philippines',
    'Qatar', 'Saudi Arabia', 'Singapore', 'Sri Lanka', 'Syrian Arab Republic',
    'Taiwan, China', 'Tajikistan', 'Thailand', 'Timor-Leste', 'Turkmenistan',
    'United Arab Emirates', 'Uzbekistan', 'Vietnam', 'Yemen, Rep.'
]

#   GDP  
asian_gdp_data = df[df['Country Name'].isin(asian_countries)]

#   (2020) GDP   
asian_gdp_data = asian_gdp_data.sort_values(by='2020', ascending=False)

#    (1990 2020)
years = list(map(str, range(1990, 2021)))

#  
plt.figure(figsize=(16, 12))
for index, row in asian_gdp_data.iterrows():
    country = row['Country Name']
    gdp_values = row[years].values.astype(float)

    # GDP      
    if not gdp_values.all():
        continue

    plt.plot(years, gdp_values, label=country)

plt.title('GDP of Asian countries (1990-2020)')
plt.xlabel('')
plt.ylabel('GDP')
plt.xticks(rotation=45)
plt.legend(loc='upper left', bbox_to_anchor=(1, 1))
plt.show()
```

```python
def get_continent(country_name):
    for continent, countries in continent_mapping.items():
        if country_name in countries:
            return continent
    return country_name  #        .
```

```python
import numpy as np
#  GDP  
gdp_data = pd.read_csv("/content/GDP Trend_all country.csv")

#    (: 'Country Name', 'Code', '1990', '2020' )
gdp_data = gdp_data[['Country Name', 'Code', '1990', '2005','2020']]
#   
continent_mapping = {
    'european_countries': european_countries,
    'north_american_countries': north_american_countries,
    'south_american_countries': south_american_countries,
    'oceania_countries': oceania_countries,
    'african_countries': african_countries,
    'asian_countries': asian_countries,
}

def get_continent(country_name):
    for continent, countries in continent_mapping.items():
        if country_name in countries:
            return continent
    return 'Other'  #       .

# GDP    
gdp_data['Continent'] = gdp_data['Country Name'].apply(get_continent)

#  GDP  
continent_gdp = gdp_data.groupby('Continent').agg({'1990': 'mean', '2020': 'mean'}).reset_index()
import matplotlib.pyplot as plt

#  
plt.figure(figsize=(14, 8))  #   

bar_width = 0.3  #    0.8  ( )

# 1990 GDP 
plt.bar(np.arange(len(continent_gdp)) - bar_width/2, continent_gdp['1990'], bar_width, label='1990 GDP', alpha=0.7)
# 2020 GDP  (     )
plt.bar(np.arange(len(continent_gdp)) + bar_width/2, continent_gdp['2020'], bar_width, label='2020 GDP', alpha=0.7)

# x   
plt.xticks(np.arange(len(continent_gdp)), continent_gdp['Continent'])  #   

plt.xlabel('Continent')
plt.ylabel('GDP')
plt.title('GDP Comparison by Continent (1990 vs. 2020)')
plt.legend()
plt.tight_layout()  #   
plt.show()
```

```python
import geopandas as gpd
import matplotlib.pyplot as plt
import pandas as pd

# 'naturalearth_lowres'  
world = gpd.read_file(gpd.datasets.get_path('naturalearth_lowres'))

# GDP   (: 'gdp_data.csv'  )
gdp_data = pd.read_csv("/content/GDP Trend_all country.csv")

# 2020 GDP  
gdp_data_2020 = gdp_data[['Code', '2020']]

# GDP     
world = world.merge(gdp_data_2020, left_on='iso_a3', right_on='Code')

#  
fig, ax = plt.subplots(1, 1, figsize=(15, 10))

#    
missing_kwds = {
    "color": "black",
    "edgecolor": "red",
    "hatch": "////",
    "label": "Missing Data"
}

#     
cmap = 'YlOrRd'  #  colormap (  )
vmin, vmax = 0, gdp_data_2020['2020'].max()  #   

world.plot(
    column='2020',
    cmap=cmap,
    linewidth=0.8,
    ax=ax,
    edgecolor='0.8',
    legend=True,
    missing_kwds=missing_kwds,  #    
    vmin=vmin, vmax=vmax  #   
)

ax.set_title('World GDP Map (2020)')

plt.show()
```
